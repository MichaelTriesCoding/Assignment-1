```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn import metrics

# Load data
cancer_df = pd.read_csv('cancer.csv')
housing_df = pd.read_csv('Housing.csv')
cancer = cancer_df.copy()
housing = housing_df.copy()
# Display the first few rows of the data
cancer.head()

# Map 'diagnosis' column to binary values
varlist = ['diagnosis']
def binary_map(x):
    return x.map({'M': 1, 'B': 0})
cancer[varlist] = cancer[varlist].apply(binary_map)

# Define features and target
X = cancer.iloc[:, 2:32].values
Y = cancer.iloc[:, 1].values

# Function to plot results
def plot_result_with_k(k_list, acc_list, precision_list, recall_list, f1score_list):
    plt.plot(k_list, acc_list, label="Accuracy")
    plt.plot(k_list, precision_list, label="Precision")
    plt.plot(k_list, recall_list, label="Recall")
    plt.plot(k_list, f1score_list, label="F1-Score")
    plt.legend()
    plt.title('Accuracy, Precision, Recall and F1-score over K values')
    plt.ylabel('Percent Value')
    plt.xlabel('K Value')
    plt.show()

# Function to calculate and print metrics
def get_results(y_test, Y_pred):
    acc = metrics.accuracy_score(y_test, Y_pred)
    pre = metrics.precision_score(y_test, Y_pred)
    rec = metrics.recall_score(y_test, Y_pred)
    fscore = metrics.f1_score(y_test, Y_pred)
    print("Accuracy:", acc)
    print("Precision:", pre)
    print("Recall:", rec)
    print("F1-Score:", fscore)
    return [acc * 100.0, pre * 100.0, rec * 100.0, fscore * 100.0]

# Function to train and evaluate model with PCA and logistic regression
def logist_model_training_pca(X, Y, classifier_type='logistic'):
    n = X.shape[1]
    acc_list, recall_list, precision_list, f1score_list, k_list = [], [], [], [], []

    for i in range(n):
        print("K =", i + 1)
        pca = PCA(n_components=i + 1)
        principalComponents = pca.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(principalComponents, Y, test_size=0.20, random_state=9999)

        if classifier_type == 'logistic':
            classifier = LogisticRegression(random_state=9)
        elif classifier_type == 'svm_poly':
            classifier = SVC(kernel='poly', C=1E2)
        elif classifier_type == 'svm_linear':
            classifier = SVC(kernel='linear', C=1E2)
        elif classifier_type == 'svm_rbf':
            classifier = SVC(kernel='rbf', C=1E2)
        else:
            raise ValueError("Invalid classifier type")

        classifier.fit(X_train, y_train)
        Y_pred = classifier.predict(X_test)
        re = get_results(y_test, Y_pred)
        acc_list.append(re[0])
        precision_list.append(re[1])
        recall_list.append(re[2])
        f1score_list.append(re[3])
        k_list.append(i + 1)

    high_acc = max(acc_list)
    high_acc_k = acc_list.index(high_acc) + 1
    print("---------------")
    print("Highest Classification Accuracy Achieved:", high_acc, "for K number =", high_acc_k)
    return k_list, acc_list, precision_list, recall_list, f1score_list

# Standardize features
sc_X = StandardScaler()
X = sc_X.fit_transform(X)

# Logistic Regression with PCA
k_list, acc_list, precision_list, recall_list, f1score_list = logist_model_training_pca(X, Y, classifier_type='logistic')
plot_result_with_k(k_list, acc_list, precision_list, recall_list, f1score_list)

# Polynomial SVM with PCA
k_list, acc_list, precision_list, recall_list, f1score_list = logist_model_training_pca(X, Y, classifier_type='svm_poly')
plot_result_with_k(k_list, acc_list, precision_list, recall_list, f1score_list)

# Linear SVM with PCA
k_list, acc_list, precision_list, recall_list, f1score_list = logist_model_training_pca(X, Y, classifier_type='svm_linear')
plot_result_with_k(k_list, acc_list, precision_list, recall_list, f1score_list)

# RBF SVM with PCA
k_list, acc_list, precision_list, recall_list, f1score_list = logist_model_training_pca(X, Y, classifier_type='svm_rbf')
plot_result_with_k(k_list, acc_list, precision_list, recall_list, f1score_list)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error

# Binary mapping of categorical features
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

# Applying the function to the housing list
housing[varlist] = housing[varlist].apply(binary_map)
housing.head()

# Define features and target variable
features = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']
X = housing.loc[:, features].values
Y = housing.loc[:, ['price']].values

# Standardize the features
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)

# Generate synthetic data for demonstration
x = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(x).ravel()
y[::5] += 3 * (0.5 - np.random.rand(8))  # Add noise to targets

# Fit regression models
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)

y_rbf = svr_rbf.fit(x, y).predict(x)
y_lin = svr_lin.fit(x, y).predict(x)
y_poly = svr_poly.fit(x, y).predict(x)

# Plot the results
lw = 2
plt.scatter(x, y, color='orange', label='data')
plt.plot(x, y_rbf, color='green', lw=lw, label='RBF model')
plt.plot(x, y_lin, color='blue', lw=lw, label='Linear model')
plt.plot(x, y_poly, color='darkblue', lw=lw, label='Polynomial model')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Support Vector Regression (SVR)')
plt.legend()
plt.show()

# Function to train and evaluate model with PCA and SVR
def linear_model_training_pca(X, Y, kernel_type='linear'):
    n = X.shape[1]
    mse_list = []
    k_list = []

    for i in range(n):
        print("Iteration Number:", i + 1)
        print("K =", i + 1)
        pca = PCA(n_components=i + 1)
        principalComponents = pca.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(principalComponents, Y, test_size=0.20, random_state=9999)

        if kernel_type == 'linear':
            classifier = SVR(kernel='linear', C=1E2)
        elif kernel_type == 'poly':
            classifier = SVR(kernel='poly', C=1E2)
        elif kernel_type == 'rbf':
            classifier = SVR(kernel='rbf', C=1E2)
        else:
            raise ValueError("Invalid kernel type")

        classifier.fit(X_train, y_train)
        Y_pred = classifier.predict(X_test)
        mse_list.append(mean_squared_error(y_test, Y_pred))
        k_list.append(i + 1)

    best_mse = min(mse_list)
    best_k = mse_list.index(best_mse) + 1
    print("----------------")
    print("Lowest MSE Achieved:", best_mse, "for K number =", best_k)

    return k_list, mse_list

# Reshape and standardize the target variable
Y = Y.reshape(-1, 1)
Y = StandardScaler().fit_transform(Y)

# Function to plot results
def plot_result_with_k(k_list, mse_scores):
    plt.plot(k_list, mse_scores, label="MSE")
    plt.legend()
    plt.title('Mean Square Error with different Ks')
    plt.ylabel('Value')
    plt.xlabel('K')
    plt.show()

# Linear SVR with PCA
k_list, mse_scores = linear_model_training_pca(X, Y, kernel_type='linear')
plot_result_with_k(k_list, mse_scores)

# Polynomial SVR with PCA
k_list, mse_scores = linear_model_training_pca(X, Y, kernel_type='poly')
plot_result_with_k(k_list, mse_scores)

# RBF SVR with PCA
k_list, mse_scores = linear_model_training_pca(X, Y, kernel_type='rbf')
plot_result_with_k(k_list, mse_scores)
```
